## Import Results

```python
# load df
df = pd.read_csv(d_path + 'train_scores.csv', sep = '|')
```

```python
# flatten score json's into df
score_keys = json.loads(df.scores[0]).keys()
for key in score_keys:
    df[key] = df.scores.apply(lambda x: json.loads(x)[key])
df.head()
```

```python
# sort results by accuracy
df['mean_test_accuracy2'] = 1 - df.mean_test_accuracy
df['mean_test_accuracy3'] = df[['mean_test_accuracy', 'mean_test_accuracy2']].max(axis = 1)
df[['model_nm', 'matrix_nm', 'params', 'mean_test_accuracy3']].sort_values(
    by = 'mean_test_accuracy3', ascending = False).head(10)
```

```python
df[['model_nm', 'matrix_nm', 'params', 'mean_test_accuracy3', 'mean_test_adjusted_rand_score']].sort_values(
    by = ['mean_test_adjusted_rand_score', 'mean_test_accuracy3'], ascending = False).head(10)
```
